```python
import pandas as pd
```

```python
df = pd.read_csv('DB.csv')
```

```python
df.drop(['Unnamed: 4','Unnamed: 5','Unnamed: 6','Unnamed: 7','Unnamed: 8','Unnamed: 9'],axis=1,inplace=True)
```

```python
df
```

```python
import pymysql
import json
from sqlalchemy import text, create_engine

engine = create_engine('mysql+pymysql://bigdata:dbgusals1@localhost:3306/chat')
ob = {}
for i in df.index:
    if pd.isna(df.iloc[i,0]):
        pass
    else:
        l = df.iloc[i,0].split(',')
        for j in l:
            j = j.replace(' ','')
                        
            if j in ob:
                ob[j] += ',' + str(i)
                
            else:
                ob[j] = str(i)

sql = """
insert into object(단어, idx)
values(
:객체명,
:jarray)
"""
for i,k in ob.items():
    객체명 = i
    jarray = k
    dt = {"객체명":객체명,"jarray":jarray}
    engine.execute(text(sql),**dt)
```
